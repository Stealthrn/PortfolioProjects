```python
# packages we will use in this project

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

%cd Downloads



# Now we need to read in the data
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
df.drop_duplicates()
```

```python
# Checking for missing data
# Looping through the data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dtypes
```

```python
#Will drop the rows with NaN values, you could do interpolation for budget and gross, I think if I did it would use rating votes and runtime to guess the budget and gross
df = df.dropna(how='any',axis = 0)
df.head()
```

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# Create correct year column
df['yearcorrect'] = df['released'].str.extract(r'(\d{4})').astype(int)
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# You can use the code below to show all rows
# pd.set_option('display.max_rows',None) 
```

```python
df_numerised = df.copy()

for col_name in df_numerised.columns:
    if(df_numerised[col_name].dtype == "object"):
        df_numerised[col_name] = df_numerised[col_name].astype("category")
        df_numerised[col_name] = df_numerised[col_name].cat.codes
        
df_numerised
```

```python
correlation_matrix = df_numerised.corr(method ='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs)
```

```python
sorted_pairs = corr_pairs.sort_values(kind="quicksort")

print(sorted_pairs)
```

```python
# We can now take a look at the ones that have a high correlation (> 0.5)

strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```

```python
# Looking at the top 15 compaies by gross revenue

CompanyGrossSum = df.groupby('company')[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values('gross', ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```

```python
CompanyGrossSum = df.groupby(['company', 'year'])[["gross"]].sum()

CompanyGrossSumSorted = CompanyGrossSum.sort_values(['gross','company','year'], ascending = False)[:15]

CompanyGrossSumSorted = CompanyGrossSumSorted['gross'].astype('int64') 

CompanyGrossSumSorted
```
